## Step A- Data Preprocessing

## Step 1: Importing libraries

```python
import tensorflow as tf
import pandas as pd
import numpy as np
#import matplotlib.pyplot as plt
```

```python
#!pip install matplotlib
```

## Step 2:Importing the dataset

```python
dataset=pd.read_csv('pima-indians-diabetes.data.csv')
```

```python
dataset
```

## Step 3: To create Feature Matrix(X) and Dependent variable vector(Y)

```python
x=dataset.iloc[:,0:8].values
y=dataset.iloc[:,8].values
```

```python
x
```

```python
y
```

## Step 6- Splitting dataset into training and testing set

```python
#!pip install sklearn
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
xtrain=sc.fit_transform(xtrain)
xtest=sc.fit_transform(xtest)
```

```python
xtrain
```

## Build Model

```python
d1=tf.keras.models.Sequential()
```

```python
#create model
#first hidden layer
d1.add(tf.keras.layers.Dense(units=12,activation='relu'))
#2nd hidden layer
d1.add(tf.keras.layers.Dense(units=8,activation='relu'))
#output layer
d1.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

## Compile the model

```python
d1.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

## Training the model

```python
d1.fit(xtrain,ytrain)
```

```python
d1.fit(xtrain, ytrain, epochs=500)
```

## Testing the model

```python
yestimated=d1.predict(xtest)
```

```python
ytest
```

```python
yestimated
```

```python
yestimated=(yestimated>0.5)
```

```python
yestimated
```

```python
print(np.concatenate((yestimated.reshape(len(yestimated),1),ytest.reshape(len(ytest),1)),1))
```

## Preformance Metrics for Neural Network

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score
cm=confusion_matrix(ytest, yestimated) 
print('confusion matrix')
print(cm)
print('accuracy=', accuracy_score(ytest,yestimated))
print('precision=', precision_score(ytest,yestimated)) 
```

```python
early_stop=EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=25)
```

## 4th model

```python
from tensorflow.keras.layers import Dropout
```

```python
#create 4th model
#first hidden layer
d1.add(tf.keras.layers.Dense(units=12,activation='relu'))
d1.add(Dropout(0.5))
#2nd hidden layer
d1.add(tf.keras.layers.Dense(units=8,activation='relu'))
d1.add(Dropout(0.5))
#output layer
d1.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

```python
d1.fit(xtrain, ytrain, epochs=500, validation_data=(xtest,ytest))
```
